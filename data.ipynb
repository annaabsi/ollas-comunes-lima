```python
import pandas as pd
```

```python
# DISTRITOS
df=pd.read_excel('distritos.xlsx')
df
```

```python
# MIDIS
df_midis=pd.read_excel('midis.xlsx', usecols=['DEPARTAMENTO', 'DISTRITO'])
df_midis=df_midis[df_midis['DEPARTAMENTO'] == 'LIMA METROPOLITANA']
df_midis=df_midis.groupby('DISTRITO').count()
df_midis.columns = ['MIDIS']
df_midis
```

```python
# MUNIDELIMA
df_munilima=pd.read_excel('munilima.xlsx', usecols=['DISTRITO','Nombre de la olla común'])
#df_munilima=df_munilima[df_munilima['DEPARTAMENTO'] == 'LIMA METROPOLITANA']
df_munilima=df_munilima.groupby('DISTRITO').count()
df_munilima.columns = ['MUNILIMA']
df_munilima
```

```python
df=df.set_index('DISTRITO').join(df_midis).join(df_munilima).fillna(0).astype(int)
df.to_csv('ollas_comunes.csv')
```
